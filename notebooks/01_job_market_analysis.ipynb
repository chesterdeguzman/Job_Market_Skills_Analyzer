# Job Market Skills Analyzer
This notebook loads the included synthetic dataset and reproduces the core portfolio analysis.

```python
import pandas as pd
df = pd.read_csv('../data/raw/job_postings_synthetic.csv', parse_dates=['posting_date'])
df.head()
```

```python
skills = df.assign(skill=df['skills'].str.split('; ')).explode('skill')
skills['skill'].value_counts().head(20)
```

```python
df.groupby('role_category').agg(postings=('job_id','count'), median_salary=('salary_mid_usd','median'), remote_share=('work_mode', lambda s: (s=='Remote').mean()))
```
